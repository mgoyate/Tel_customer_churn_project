# AUC AND ROC SCORE

```python
from sklearn.metrics import roc_auc_score
import sklearn.metrics as metrics
```

```python
fpr,tpr,threshold = metrics.roc_curve(ytest,ypred)
```

```python
roc_auc = metrics.auc(fpr,tpr)
```

```python
roc_auc
```

```python
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# decision tree 

```python
from sklearn.tree import DecisionTreeClassifier#for checking testing results
from sklearn.metrics import classification_report, confusion_matrix#for visualizing tree 
from sklearn.tree import tree
```

```python
x= df2.values[:,1:5]
y=df2.values[:,0]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=100) 
```

```python
clf_entropy = DecisionTreeClassifier(criterion="entropy",random_state=100,max_depth=3,min_samples_leaf=5)
clf_entropy.fit(xtrain,ytrain)
```

```python
y_pred_en = clf_entropy.predict(xtest)
```

```python
y_pred_en
```

```python
print(("accuracy is "),accuracy_score(ytest,y_pred_en)*100)
```
